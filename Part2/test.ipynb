```python
def adder32bit_gen(file_handler):
    file_handler.write('P0 := AND(AND(A0,AND(A0,B0)), AND(AND(A0,B0),B0));\n')
    file_handler.write('G0 := AND(A0,B0);\n')
    file_handler.write('S0 := S0 := AND(AND(P0,AND(P0,ALUFN[0])), AND(AND(P0,0),ALUFN[0]));\n')
    file_handler.write('\n')

    #for the pg
    for i in range(1, 32):
        currentPi = 'P{0} := AND(AND(A{0},AND(A{0},B{0})), AND(AND(A{0},B{0}),B{0}));\n'.format(i)
        currentGi = 'G{0} := AND(A{0},B{0});\n'.format(i)
        file_handler.write(currentPi + currentGi + '\n')
		#currentSum = 'S{0} := ODD(A{0},B{0},C{1});\n'.format(i, i-1)
		#currentSum = 'S{0} := ODD(A{0},B{0},C{1});\n'.format(i, i)
		#currentCarryOut = 'C{0} := OR(AND(A{0},B{0}),AND(C{1},A{0}),AND(C{1},B{0}));\n'.format(i, i-1)
		#currentCarryOut = 'C{1} := OR(AND(A{0},B{0}),AND(C{0},A{0}),AND(C{0},B{0}));\n'.format(i, i+1)

    #for layer 1
    for i in range(1, 32):
        currentPi = 'P1{0} := AND(P{1},P{2});\n'.format(i+1, i, i-1)
        currentGi = 'G1{0} := OR(AND(P{1},G{2}),G{1};\n'.format(i+1, i, i-1)
        file_handler.write(currentPi + currentGi + '\n')
    file_handler.write("C0 := OR(AND(P0,ALUFN[0]),G0);\n")

    #for layer 2
    for i in range(5, 32):
        currentPi = 'P2{0} := AND(P1{1},P1{2});\n'.format(i+1, i, i-1)
        currentGi = 'G2{0} := OR(AND(P1{1},G1{2}),G1{1};\n'.format(i+1, i, i-1)
        file_handler.write(currentPi + currentGi + '\n')
    file_handler.write("C2 := OR(AND(P12,C0),G12);\n")
    file_handler.write("C1 := OR(AND(P11,ALUFN[0]),G11);\n")

    #for layer 3
    for i in range(9, 32):
        currentPi = 'P3{0} := AND(P2{1},P2{2});\n'.format(i+1, i, i-1)
        currentGi = 'G3{0} := OR(AND(P2{1},G2{2}),G2{1};\n'.format(i+1, i, i-1)
        file_handler.write(currentPi + currentGi + '\n')
    for j in range(4,7):
        ksg1 = "C{0} := OR(AND(P2{0},C0),G2{j});\n".format(j)
        ksg2 = "C{0} := OR(AND(P2{0},C1),G2{j});\n".format(j)
        ksg3 = "C{0} := OR(AND(P2{0},C2),G2{j});\n".format(j)
        file_handler.write(ksg1+ksg2+ksg3 + '\n')
    file_handler.write("C3 := OR(AND(P23,ALUFN[0]),G23);\n")

    #for layer 4
    for i in range(17, 32):
        currentPi = 'P4{0} := AND(P3{1},P3{2});\n'.format(i+1, i, i-1)
        currentGi = 'G4{0} := OR(AND(P3{1},G3{2}),G3{1};\n'.format(i+1, i, i-1)
        file_handler.write(currentPi + currentGi + '\n')
    for j in range(8,15):
        ksg1 = "C{0} := OR(AND(P2{0},C0),G2{j});\n".format(j)
        ksg2 = "C{0} := OR(AND(P2{0},C1),G2{j});\n".format(j)
        ksg3 = "C{0} := OR(AND(P2{0},C2),G2{j});\n".format(j)
        ksg4 = "C{0} := OR(AND(P2{0},C3),G2{j});\n".format(j)
        ksg5 = "C{0} := OR(AND(P2{0},C4),G2{j});\n".format(j)
        ksg6 = "C{0} := OR(AND(P2{0},C5),G2{j});\n".format(j)
        ksg7 = "C{0} := OR(AND(P2{0},C6),G2{j});\n".format(j)
        file_handler.write(ksg1+ksg2+ksg3+ksg4+ksg5+ksg6+ksg7 + '\n')
    file_handler.write("C7 := OR(AND(P37,ALUFN[0]),G37);\n")
		




if __name__=='__main__':
	bcfile = open('adder32bit.bc','w')
	bcfile.write('BC1.1\n')
	adder32bit_gen(bcfile)
	bcfile.close()
```
